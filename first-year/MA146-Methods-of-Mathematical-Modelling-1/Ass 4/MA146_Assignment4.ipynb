```python
from sympy import Function, Symbol, Eq, symbols, Derivative, dsolve, pi 
from sympy.plotting import plot, plot_parametric
```

Solving a linear system of first order differential equations with constant coefficients of the form 
$$ \binom{x_1'(t)}{x_2'(t)} = \begin{pmatrix} 0 & 1 \\ -c & -b \end{pmatrix} \binom{x_1(t)}{x_2(t)}$$
with numbers $b,c \in \mathbb{R}$. 

The system is implemented using the ${\tt Eq}$ function (second row below), which sets the two arguments equal to each other and returns an equation. The solver function ${\tt dsolve}$ also accepts systems.  

```python
b, c = 0, 16
x_1, x_2 = symbols("x_1 x_2", cls=Function)
t = Symbol('t')
ode_sys = [Eq(x_1(t).diff(t), x_2(t)), Eq(x_2(t).diff(t), -c*x_1(t) -b*x_2(t))]
dsolve(ode_sys)
```

The output means that 
\begin{align*}
 x_1(t) &= \tfrac{1}{4} C_1 \sin(4t) + \tfrac{1}{4} C_2 \cos(4t), \\
 x_2(t) &= C_1 \cos(4t) - C_2 \sin(4t),
\end{align*}
for any $C_1,C_2 \in \mathbb{R}$. 

Solving the system with initial conditions $x_1(\pi) = 13$, $x_2(\pi) = 8$ and then plotting the curve parameterised by the two components of the solution over time in the $x_1$-$x_2$ plane. 

```python
sol_sys = dsolve(ode_sys, ics={x_1(pi): 13, x_2(pi): 8})
print(sol_sys)
plot_parametric((sol_sys[0].rhs,sol_sys[1].rhs),(t,0,4*pi),label='$(x_1(t),x_2(t))$',xlabel='$x_1$',ylabel='$x_2$',legend=True)
```
